### Learning curve

```python
def plot_learningcurve(phases, all_losses):
    """
    all_losses can be an array or a matrix
    """
    f, ax = plt.subplots(figsize=(6, 4))
    ax.spines['top'].set_visible(False)  #
    ax.spines['right'].set_visible(False) #
    ax.set_xlabel('Trials',labelpad = 7, fontdict={'weight': 'normal', 'size': 13})
    ax.set_ylabel('Losses',labelpad = 7, fontdict={'weight': 'normal', 'size': 13})
    ax.set_title("Learning Curve",fontdict={'weight': 'normal', 'size': 15})
    for j in range(len(phases)):
        all_loss = all_losses[j]
        for i in range(len(all_loss)):
            if all_loss[i] < 0.1:
                ax.text(i*0.9, all_loss[i]*1.3, s=i)
                break
        ax.plot(all_loss, label=phases[j])
        ax.legend(bbox_to_anchor=(1.3, 0.25))
    plt.savefig("learning_curve_for_all_losses.png",quality=95, dpi=800,bbox_inches='tight')    
plot_learningcurve(['Phase A', 'Phase B', 'Phase C'], 
                   [all_lossesA, all_lossesB, all_lossesC])
```

```python
def plot_learning_accu_curve(phases, all_losses):
    """
    all_losses can be an array or a matrix
    """
    f, ax = plt.subplots(figsize=(6, 4))
    ax.spines['top'].set_visible(False)  #
    ax.spines['right'].set_visible(False) #
    ax.set_xlabel('Trials',labelpad = 7, fontdict={'weight': 'normal', 'size': 13})
    ax.set_ylabel('Accuracy',labelpad = 7, fontdict={'weight': 'normal', 'size': 13})
    ax.set_title("Learning Curve",fontdict={'weight': 'normal', 'size': 15})
    for j in range(len(phases)):
        all_loss = all_losses[j]
        for i in range(len(all_loss)):
            if all_loss[i] > 0.9:
                ax.text(i*0.9, all_loss[i], s=i)
                break
        ax.plot(all_loss, label=phases[j])
        ax.legend(bbox_to_anchor=(1.3, 0.25))
    plt.savefig("learning_curve_for_all_accu.png",quality=95, dpi=800,bbox_inches='tight')    
plot_learning_accu_curve(['Phase A', 'Phase B', 'Phase C'], 
                  [all_accuAA, all_accuBB, all_accuCC])
```
